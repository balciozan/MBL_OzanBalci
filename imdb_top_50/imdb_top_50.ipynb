```python
# MBL 549E - MBL 549E - Special Top.in Arch.Des.Comp.
# Week 4 Homework
# Web Scrapping
#
# Ozan BalcÄ± - 020130369

import bs4
import pandas as pd
import requests
```

```python
response = requests.get('https://www.imdb.com/search/title/?groups=top_250&sort=user_rating')
print(response.text) 
```

```python
soup = bs4.BeautifulSoup(response.text, "html.parser")
print(soup.prettify()) # Print the output using the 'prettify' function
```

```python
soup.title
```

```python
soup.title.string
```

```python
data = soup.findAll (attrs={'class': 'lister-item-index unbold text-primary'})
for i in data: # Numbers
    print(i.string)
```

```python
movies = soup.findAll('h3', class_='lister-item-header')
data2 = [movie.find('a') for movie in movies]
for i in data2: # Movie Names
    print(i.string)
```

```python
data3 = soup.findAll (attrs={'class': 'runtime'})
for i in data3: # Durations
    print(i.string)
```

```python
data4 = soup.findAll (attrs={'class': 'genre'})
for i in data4: # Genres
    print(i.string)
```

```python
data5 = soup.findAll (attrs={'class': 'lister-item-year text-muted unbold'})

for i in data5: # Years
    print(i.string)

    
```
